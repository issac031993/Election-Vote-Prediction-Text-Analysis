```python
import nltk
from nltk.corpus import inaugural

nltk.download('inaugural')

# Loading the inaugural speeches
roosevelt_speech = inaugural.raw('1941-Roosevelt.txt')
kennedy_speech = inaugural.raw('1961-Kennedy.txt')
nixon_speech = inaugural.raw('1973-Nixon.txt')

# Function to get statistics
def get_stats(text):
    num_characters = len(text)
    num_words = len(nltk.word_tokenize(text))
    num_sentences = len(nltk.sent_tokenize(text))
    return num_characters, num_words, num_sentences

# statistics for each speech
roosevelt_stats = get_stats(roosevelt_speech)
kennedy_stats = get_stats(kennedy_speech)
nixon_stats = get_stats(nixon_speech)

# Printing the results
print("2.1 Number of characters, words, and sentences:")
print("Roosevelt:", roosevelt_stats)
print("Kennedy:", kennedy_stats)
print("Nixon:", nixon_stats)
```

```python
import nltk
from nltk.corpus import inaugural
from nltk.corpus import stopwords

nltk.download('inaugural')
nltk.download('stopwords')

# Loading the inaugural speeches
roosevelt_speech = inaugural.raw('1941-Roosevelt.txt')
kennedy_speech = inaugural.raw('1961-Kennedy.txt')
nixon_speech = inaugural.raw('1973-Nixon.txt')

# Loading stopwords
stop_words = set(stopwords.words('english'))

# Function to remove stopwords and show statistics
def remove_stopwords_and_show_stats(text, president_name):
    words_before_removal = len(nltk.word_tokenize(text))
    
    # Remove stopwords
    words = nltk.word_tokenize(text)
    filtered_words = [word for word in words if word.lower() not in stop_words]
    
    # Showing sample sentence after removal of stopwords
    sample_sentence = ' '.join(filtered_words[:10])  # Displaying the first 10 words as a sample
    print(f"Sample sentence after removing stopwords for {president_name}: {sample_sentence}")
    
    words_after_removal = len(filtered_words)
    
    return words_before_removal, words_after_removal, filtered_words

# Removing stopwords for each speech and showing statistics
roosevelt_stats_stopwords = remove_stopwords_and_show_stats(roosevelt_speech, "Roosevelt")
kennedy_stats_stopwords = remove_stopwords_and_show_stats(kennedy_speech, "Kennedy")
nixon_stats_stopwords = remove_stopwords_and_show_stats(nixon_speech, "Nixon")

# Printing the results
print("\n2.2 Word count before and after removing stopwords:")
print("Roosevelt:", roosevelt_stats_stopwords[0], "->", roosevelt_stats_stopwords[1])
print("Kennedy:", kennedy_stats_stopwords[0], "->", kennedy_stats_stopwords[1])
print("Nixon:", nixon_stats_stopwords[0], "->", nixon_stats_stopwords[1])
```

```python
import nltk
from nltk.corpus import inaugural
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import re  # Import the regular expression library

nltk.download('inaugural')
nltk.download('stopwords')

# Loading the inaugural speeches
roosevelt_speech = inaugural.raw('1941-Roosevelt.txt')
kennedy_speech = inaugural.raw('1961-Kennedy.txt')
nixon_speech = inaugural.raw('1973-Nixon.txt')

# Loading stopwords
stop_words = set(stopwords.words('english'))

# Function to clean text (remove hyphens, non-alphabetic characters, and numbers) and get most common words
def clean_and_get_most_common_words(text, top_n=3):
    # Remove hyphens, non-alphabetic characters, and numbers
    cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Tokenize and remove stopwords
    words = nltk.word_tokenize(cleaned_text)
    filtered_words = [word for word in words if word.lower() not in stop_words and word.isalpha()]
    
    # Get most common words
    freq_dist = FreqDist(filtered_words)
    common_words = freq_dist.most_common(top_n)
    
    return common_words

# Getting the most common words for each president after cleaning
roosevelt_common_words_cleaned = clean_and_get_most_common_words(roosevelt_speech)
kennedy_common_words_cleaned = clean_and_get_most_common_words(kennedy_speech)
nixon_common_words_cleaned = clean_and_get_most_common_words(nixon_speech)

# Printing the results
print("\n2.3 Most common words after removing stopwords, hyphens, non-alphabetic characters, and numbers:")
print("Roosevelt:", roosevelt_common_words_cleaned)
print("Kennedy:", kennedy_common_words_cleaned)
print("Nixon:", nixon_common_words_cleaned)
```

```python
import nltk
from nltk.corpus import inaugural
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt

nltk.download('inaugural')
nltk.download('stopwords')

# Load the inaugural speeches
roosevelt_speech = inaugural.raw('1941-Roosevelt.txt')
kennedy_speech = inaugural.raw('1961-Kennedy.txt')
nixon_speech = inaugural.raw('1973-Nixon.txt')

# Loading stopwords
stop_words = set(stopwords.words('english'))

# Function to plot word cloud
def plot_word_cloud(text, title):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(title)
    plt.show()

# Removing stopwords and plot word cloud for each president
roosevelt_no_stopwords = ' '.join([word for word in nltk.word_tokenize(roosevelt_speech) if word.lower() not in stop_words])
kennedy_no_stopwords = ' '.join([word for word in nltk.word_tokenize(kennedy_speech) if word.lower() not in stop_words])
nixon_no_stopwords = ' '.join([word for word in nltk.word_tokenize(nixon_speech) if word.lower() not in stop_words])

plot_word_cloud(roosevelt_no_stopwords, "Roosevelt Inaugural Speech Word Cloud")
plot_word_cloud(kennedy_no_stopwords, "Kennedy Inaugural Speech Word Cloud")
plot_word_cloud(nixon_no_stopwords, "Nixon Inaugural Speech Word Cloud")
```
